```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
with open('unicycle_exp.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
states = np.concatenate(data["states"]).reshape(-1, 3)
p_des = np.concatenate(data["p_des_list"]).reshape(-1, 2)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))

ax.plot(states[:, 0], states[:, 1], color="cornflowerblue", linewidth=3)
ax.plot(p_des[:, 0], p_des[:, 1], "o", color='darkorange', markersize=10)

plt.show()
```
