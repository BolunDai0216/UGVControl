```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import pickle

mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'
plt.rcParams["font.size"] = 16
```

```python
with open('data/unicycle_exp.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
states = np.concatenate(data["states"]).reshape(-1, 3)
p_des = np.concatenate(data["p_des_list"]).reshape(-1, 2)
controls = np.concatenate(data["controls"]).reshape(-1, 2)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))

ax.plot(states[:, 0], states[:, 1], color="cornflowerblue", linewidth=3, zorder=20, label="Trajectory")
ax.plot(p_des[:, 0], p_des[:, 1], "o", color='darkorange', markersize=5, zorder=10, label="Waypoints")

ax.xaxis.set_major_locator(MultipleLocator(2.0))
ax.xaxis.set_minor_locator(MultipleLocator(2.0/4))
ax.yaxis.set_major_locator(MultipleLocator(2.0))
ax.yaxis.set_minor_locator(MultipleLocator(2.0/4))

ax.grid(True, "minor", color="0.85", linewidth=0.50, zorder=-20)
ax.grid(True, "major", color="0.65", linewidth=0.75, zorder=-10)
ax.tick_params(which="both", bottom=False, left=False)

ax.set_xlim([10.0, 21])
ax.set_ylim([6.0, 8.5])
ax.set_aspect('equal')
ax.set_xlabel("X [m]")
ax.set_ylabel("Y [m]")
ax.legend(frameon=False, ncol=2, loc="upper center", bbox_to_anchor=(0.5, 1.6))

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 3))

axs[0].plot(controls[:, 0], color="cornflowerblue", linewidth=3)
axs[0].set_title("Linear Velocity [m/s]")

axs[1].plot(controls[:, 1], color="cornflowerblue", linewidth=3)
axs[1].set_title("Angular Velocity [rad/s]")

plt.show()
```
