```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
with open('unicycle_exp.pkl', 'rb') as f:
    states = pickle.load(f)
```

```python
states = np.concatenate(states).reshape(-1, 3)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))

ax.plot(states[:, 0], states[:, 1], color="cornflowerblue", linewidth=3)
ax.plot([2], [2], "o", color='darkorange', markersize=10)

plt.show()
```
