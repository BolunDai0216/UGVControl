```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
with open('data/unicycle_exp.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
states = np.concatenate(data["states"]).reshape(-1, 3)
p_des = np.concatenate(data["p_des_list"]).reshape(-1, 2)
controls = np.concatenate(data["controls"]).reshape(-1, 2)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))

ax.plot(states[:, 0], states[:, 1], color="cornflowerblue", linewidth=3, zorder=10)
ax.plot(p_des[:, 0], p_des[:, 1], "o", color='darkorange', markersize=5, zorder=-10)

ax.set_xlim([10.0, 21])
ax.set_ylim([5.0, 8.5])
ax.set_aspect('equal')

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].plot(controls[:, 0], color="cornflowerblue", linewidth=3)
axs[1].plot(controls[:, 1], color="cornflowerblue", linewidth=3)
```
